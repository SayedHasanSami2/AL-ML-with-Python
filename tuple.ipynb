```python

def safe_divide(a,b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero"
    
nmbrs = list(map(int,input().split()))
ans = safe_divide(nmbrs[0], nmbrs[1])
print(ans)
```

```python
import math

class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return math.pi * self.r ** 2

taking_inputs = input().split()
frist_part = taking_inputs[0]
if frist_part == "Rectangle":
    w = float(taking_inputs[1])
    h = float(taking_inputs[2])
    shape = Rectangle(w, h)
elif frist_part == "Circle":
    r = float(taking_inputs[1])
    shape = Circle(r)
else:
    shape = Shape()

print("Area: {:.2f}".format(shape.area()))
```

```python
import math

nmbr = int(input())
OG = nmbr
sum_of_fcrial = 0

while nmbr > 0:
    digit = nmbr % 10
    sum_of_fcrial += math.factorial(digit)
    nmbr //= 10

if sum_of_fcrial == OG:
    print("Strong Number")
else:
    print("Not Strong Number")
```
