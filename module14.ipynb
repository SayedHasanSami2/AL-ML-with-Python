```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
    "id": ids,
    "age": ages,
    "salary": salaries,
    "dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
print(DF)
```

Scale the value

```python
##Scale the value of salary column between 0 and 1

mn = DF['salary'].min()
mx = DF['salary'].max()
DF['Scaled_Salary'] = (DF['salary'].apply(lambda x: (x - mn) / (mx - mn)))
print(DF)
```
