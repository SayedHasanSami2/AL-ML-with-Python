```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Use built-in seaborn dataset
student = sns.load_dataset('tips')
student.head()
```

```python
#axes level
sns.lineplot(data= student , x = 'size' , y = 'total_bill',errorbar=None)
```

```python
sns.lineplot(data= student , x = 'size' , y = 'total_bill',errorbar=None,hue='sex')
```

```python
#figure level

sns.relplot(kind ='line', data= student , x = 'size' , y = 'total_bill',errorbar=None,hue='sex')
```

```python
sns.relplot(kind ='line', data= student , x = 'size' , y = 'tip',errorbar=None,hue='sex')
```

```python
sns.relplot(kind ='line', data= student , x = 'size' , y = 'tip',errorbar=None,hue='time')
```

Scatter plot using Seaborn

```python
sns.scatterplot(data = student , x = 'total_bill' , y ='tip') 
```

```python
sns.scatterplot(data = student , x = 'total_bill' , y ='tip',hue='sex', style='time')
```

```python
sns.scatterplot(data = student , x = 'total_bill' , y ='tip',hue='sex', style='smoker')
```

```python
sns.scatterplot(data = student , x = 'total_bill' , y ='tip',hue='sex', style='smoker', size='size')
```

```python
sns.relplot(kind='scatter', data = student , x = 'total_bill' , y ='tip',hue='sex',style='smoker',size='size')
```
