# Question 2 (20 Marks)
## Handle missing values: Fill missing Salary with department-wise median
## Calculate total Salary sum for employees with ExperienceYears > 12 and ProjectCount > 5

**Final output:** One numeric value

```python
# Load dataset
df = pd.read_csv('final-employee-ds.csv')
print("Dataset loaded successfully")
```

```python
# Handle missing values: Fill missing Salary with department-wise median
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))
print("Missing values handled")
```

```python
# Calculate total Salary sum for employees with ExperienceYears > 12 and ProjectCount > 5
filtered_employees = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]
total_salary_sum = filtered_employees['Salary'].sum()

# Final output: One numeric value
print(total_salary_sum)
```

# Question 3 (20 Marks)
## Filter employees: Department = "IT" or "Finance", Age between 30–45, PerformanceScore > 88
## Sort by Salary descending

**Final output:** Table with Name, Department, Salary, PerformanceScore

```python
# Load dataset
df = pd.read_csv('final-employee-ds.csv')
```

```python
# Filter employees: Department = "IT" or "Finance"
dept_filter = df['Department'].isin(['IT', 'Finance'])
```

```python
# Age between 30–45
age_filter = (df['Age'] >= 30) & (df['Age'] <= 45)
```

```python
# PerformanceScore > 88
performance_filter = df['PerformanceScore'] > 88
```

```python
# Apply all filters and sort by Salary descending
filtered_employees = df[dept_filter & age_filter & performance_filter]
result_table = filtered_employees[['Name', 'Department', 'Salary', 'PerformanceScore']].sort_values('Salary', ascending=False)

# Final output: Table with Name, Department, Salary, PerformanceScore
print(result_table)
```

# Question 4 (20 Marks)
## Add new column SalaryPerHour = Salary ÷ (WorkHoursPerWeek × 4)
## Filter top 5 employees by SalaryPerHour

**Final output:** Table(5 Rows) with Name, SalaryPerHour

```python
# Load dataset
df = pd.read_csv('final-employee-ds.csv')
```

```python
# Add new column SalaryPerHour = Salary ÷ (WorkHoursPerWeek × 4)
df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)
```

```python
# Filter top 5 employees by SalaryPerHour
top_5_employees = df.nlargest(5, 'SalaryPerHour')[['Name', 'SalaryPerHour']]

# Final output: Table(5 Rows) with Name, SalaryPerHour
print(top_5_employees)
```

## Create a bar chart showing the count of employees in each Department, separated by Gender

**Dataset Columns Used:**
- Department → Categorical variable (IT, Finance, HR, Marketing, Management)
- Gender → Categorical variable (Male, Female)

**Expected Output:** Bar chart with x-axis: Department, y-axis: Number of employees, bars split by Gender with different colors

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import libraries and load dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('final-employee-ds.csv')

# Create bar chart showing count of employees in each Department, separated by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Gender', palette=['skyblue', 'salmon'])
plt.title('Count of Employees by Department and Gender')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
