```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
hours = [2 ,3 ,4 , 1 , 5 , 7 , 3]
days = [ 1, 2 , 3 ,4 ,5 ,6 ,7 ]
plt.plot(days, hours)
plt.title("Study Hours Over Days")
# plt.show()
```

```python
hours_1 = [2 ,3 ,4 , 1 , 5 , 7 , 3]
hours_2 = [1 ,4 ,2 , 5 , 6 , 2 , 4]
days = [ 1, 2 , 3 ,4 ,5 ,6 ,7 ]
plt.plot(days, hours_1)
plt.plot(days, hours_2)
plt.title("Study Hours Over Days")
# plt.show()
```

2D line plot from Pandas Data


```python
df = pd.read_csv('employees.csv')
print(df.head())
```

```python
plt.plot(df['id'], df['salary'])
plt.plot(df['id'], df['age'])
plt.title("ID vs Salary")
```

```python
# Better visualization with separate plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot 1: ID vs Salary
ax1.plot(df['id'], df['salary'], 'b-o')
ax1.set_title('ID vs Salary')
ax1.set_xlabel('ID')
ax1.set_ylabel('Salary')

# Plot 2: ID vs Age
ax2.plot(df['id'], df['age'], 'r-o')
ax2.set_title('ID vs Age')
ax2.set_xlabel('ID')
ax2.set_ylabel('Age')

plt.tight_layout()
plt.show()
```
