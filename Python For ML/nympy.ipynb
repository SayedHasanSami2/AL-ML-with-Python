```python
import numpy as np

arr1 = np.array([1, 2, 3])
print(arr1)

arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2)
```

```python
import numpy as ns

arr = ns.random.rand(3,4)
print(arr)

arr2 = ns.random.randint(1,10,size=(3,4))
print(arr2)

diagonal_matrix = ns.diag([1,2,3,4])
print(diagonal_matrix)

identity_matrix = ns.eye(3)
print(identity_matrix)
```

```python
###Transpose of a matrix

import numpy as np

# Define a matrix
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Transpose the matrix
transpose_matrix = matrix.T

print("Original matrix:")
print(matrix)

print("\nTransposed matrix:")
print(transpose_matrix)
```

```python
### array Splitting
import numpy as np
array = np.random.randint(1, 20, size=(4, 4))
print("Original Array:")
print(array)
split_arrays = np.array_split(array, 3)
print("\nSplit Arrays:") 
print(split_arrays)
# split is used when the array can be evenly divided
even_split_arrays = np.split(array, 2)
print("\nEven Split Arrays:")
print(even_split_arrays)
```
